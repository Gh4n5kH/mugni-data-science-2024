Q2 - Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%)
per kolom?

```python
missing = df.isnull().sum()
percent = (missing / len(df)) * 100

pd.DataFrame({
    'Missing': missing,
    'Percent (%)': percent
})
```

**jawaban:**

Terdapat 3 missing values:

* Age - 177 - 19,8%

* Cabin - 687 - 77,1%

* Embarked - 2 - 0,22%

Q3 - Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy
atau Pandas.

```python
survival_rate = df['Survived'].mean() * 100
survival_rate
```

**Jawaban:**

Tingkat keselamatan penumpang Titanic 38.38%
Sebaliknya, 61.62% tidak selamat dalam tragedi tersebut.

Q4 - Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-
rata usianya?

```python
female_p1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

count = female_p1.shape[0]
avg_age = female_p1['Age'].mean()

count, avg_age
```

**Jawaban:**

Jumlahnya 94 orang dengan rata-rata usia 34,61 tahun.

Q5 - Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas
penumpang. Kelas mana yang paling tinggi?

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean()
survival_by_class
```

**Jawaban:**

Penumpang kelas 1 memiliki tingkat keselamatan tertinggi, yaitu 62,96%.
Semakin rendah kelas penumpang, semakin kecil peluang untuk selamat
