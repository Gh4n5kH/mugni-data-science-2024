# ── TEMPLATE DASHBOARD VISUALISASI STATIS ──

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import seaborn as sns
```

```python
# ── STYLE VISUAL ──
sns.set_theme(style="whitegrid")

# ── 1. LOAD & INSPECT DATASET ──
df = sns.load_dataset('tips')

print("Shape Dataset:")
print(df.shape)

print("\nTipe Data:")
print(df.dtypes)

print("\nStatistik Deskriptif:")
print(df.describe().round(2))

# ── 2. SETUP FIGURE LAYOUT ──
fig = plt.figure(figsize=(16, 12))

fig.suptitle(
    'Dashboard Analisis Dataset Tips',
    fontsize=18,
    fontweight='bold',
    y=0.98
)

# Grid layout 2x2
gs = gridspec.GridSpec(
    2, 2,
    figure=fig,
    hspace=0.4,
    wspace=0.35
)

# Membuat subplot
ax1 = fig.add_subplot(gs[0, 0])  # kiri atas
ax2 = fig.add_subplot(gs[0, 1])  # kanan atas
ax3 = fig.add_subplot(gs[1, 0])  # kiri bawah
ax4 = fig.add_subplot(gs[1, 1])  # kanan bawah

# ── 3. GRAFIK 1: BAR CHART ──
avg_by_day = df.groupby('day')['total_bill'].mean()

avg_by_day.plot(
    kind='bar',
    ax=ax1,
    color='#028090',
    edgecolor='white'
)

ax1.set_title('Rata-rata Total Tagihan per Hari')
ax1.set_xlabel('Hari')
ax1.set_ylabel('Total Bill (USD)')

ax1.tick_params(axis='x', rotation=0)

ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# ── 4. GRAFIK 2: HISTOGRAM + KDE ──
sns.histplot(
    data=df,
    x='total_bill',
    hue='time',
    kde=True,
    palette='Set2',
    ax=ax2
)

ax2.set_title('Distribusi Total Tagihan per Waktu Makan')

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

# ── 5. GRAFIK 3: BOXPLOT ──
sns.boxplot(
    data=df,
    x='day',
    y='total_bill',
    hue='sex',
    palette='Set1',
    ax=ax3
)

ax3.set_title('Distribusi Tagihan per Hari & Gender')

ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)

# ── 6. GRAFIK 4: SCATTER PLOT ──
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    size='size',
    sizes=(30, 180),
    palette='Set2',
    alpha=0.7,
    ax=ax4
)

# Garis regresi
sns.regplot(
    data=df,
    x='total_bill',
    y='tip',
    scatter=False,
    color='gray',
    ax=ax4
)

ax4.set_title('Hubungan Tagihan & Tip (Ukuran = Jumlah Tamu)')

ax4.spines['top'].set_visible(False)
ax4.spines['right'].set_visible(False)

# ── 7. EXPORT & SHOW ──
plt.savefig(
    'dashboard_tips.png',
    dpi=150,
    bbox_inches='tight'
)

plt.show()

print("\nDashboard tersimpan sebagai 'dashboard_tips.png'")
```
